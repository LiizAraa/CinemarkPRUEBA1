## Objetivo

Leer 10 notas de estudiantes y calcular:

- Promedio
- Nota máxima
- Nota mínima
- Cantidad de aprobados


```python
aprobados = 0
suma = 0

for i in range(10):
    nota = float(input(f"Ingrese la nota {i+1}: "))


```

/INICIAMOS CON LA LISTA Y LA SUMA DE LAS NOTAS EN CERO/
EL VALOR i en la lista de (10)
en nota abrimos la variable *"introduzca nota"*



```python
aprobados = 0
suma = 0

for i in range(10):
    nota = float(input(f"Ingrese la nota {i+1}: "))

    suma += nota

    if i == 0:
        maximo = nota
        minimo = nota

    if nota > maximo:
        maximo = nota

    if nota < minimo:
        minimo = nota

    if nota >= 4:
        aprobados += 1

promedio = suma / 10

print("Cantidad de aprobados:", aprobados)
print("Promedio:", promedio)
print("Máximo:", maximo)
print("Mínimo:", minimo)
```

Agregando una nota, si la nota es cero el maximo y minimo sera cero.
Si nota es mayor o igual a 4 suma en aprobados + 1;
luego de sumar todas las notas se saca promedio
